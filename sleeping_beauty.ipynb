```python
import numpy as np
import matplotlib.pyplot as plt
import time

def rng():
    return np.random.default_rng(int(1486159*time.time()))

heads = 0
tails = 0
h1 = []
t1 = []
for i in range(1,10000):
    coin_toss = np.random.randint(1,3)
    if coin_toss == 1:
        heads += 1
        h1.append([i, heads/(heads+tails)])
    else:
        tails += 2
        t1.append([i, tails/(heads+tails)])
    
h1 = np.array(h1)
t1 = np.array(t1)

plt.plot(h1[:,0], h1[:,1], 'r-')
plt.plot(t1[:,0], t1[:,1], 'b-')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import time

fig, ax = plt.subplots()

def rng():
    return np.random.default_rng(int(1486159*time.time()))

heads = 0
tails = 0
h1 = []
t1 = []

for i in range(1, 100000):
    coin_toss = np.random.randint(1, 3)
    if coin_toss == 1:
        heads += 1
        h1.append([i, heads/(heads+tails)])
    else:
        tails += 2
        t1.append([i, tails/(heads+tails)])

h1 = np.array(h1)
t1 = np.array(t1)

h = ax.plot(h1[:,0], h1[:,1], 'r-', label='P(Heads)')
t = ax.plot(t1[:,0], t1[:,1], 'b-', label='P(Tails)')

ax.legend()
ax.set_xlabel('Experiment number')
ax.set_ylabel('Probability')

plt.show()
```

```python
%%time
import seaborn as sns
import time
import itertools
import numpy as np
import matplotlib.pyplot as plt

def rng():
    return np.random.default_rng(int(1486159*time.time()))

mean = 0
std_deviation = 1
n_samples = 100000

x = np.linspace(-20,20,10000)
l = (1/(std_deviation*np.sqrt(2*np.pi)))*(np.e)**(-(((x-mean)/std_deviation)**2)/2)
n = []

for i in range(len(x)-1):
    n.append(list(np.linspace(x[i],x[i+1],int(l[i]*10000))))
n = list(itertools.chain(*n))
o = (rng().choice(n, 100000))
#plt.hist(o, bins = 60)
sns.displot(data=o,kind='ecdf')
plt.show(block=False)
```
